## What breed of dogs are the least likely to be spayed? Male or female?

HINTS FOR: "Which dogs are least likely to be spayed or neutere
To calculate the percent of unfixed animals, you need to do three thing
1. Find the number of dogs of each breed
2. Find the number of dogs of each breed that are NOT fixed.
3. Divide the second number by the first number, and then sort it.

That's it. That's all you need to do. Do those, step by step, and you'll get it.

Tip: Check your work by making sure one of your 100% unfixed dogs is a Mudi. It might not be the first one in your list, though, since there are a bunch of 100% unfixed ones (small sample size!).

## Which borough has the highest number of dogs per-capita?

Youâ€™ll need to merge in `population_boro.csv`

HINTS FOR: "Which borough has the highest number of dogs per capita?"

You just need to divide the number of dogs by the number of people. You knew that, though. Here are your steps:

1. Calculate the number of dogs in each borough.

2. Calculate the number of people in each borough.

3. Divide the first number by the second number. 

I know you are looking at me like I am crazy, but... it's true.

Tip: Check your work with Staten Island being about 0.0057.

Tip: If you figured out the one up above, it's kind of similar in terms of unexpected math things.

Tip: What's the median of [2, 2, 2, 2, 2, 2, 2, 2, 2]? What's the median of [105, 105, 105, 105, 105, 105, 105]? What's the median of 2,727 instances of the number 474,558?

Tip: You use groupby for one of these, but not both.

```python
df_pop = pd.read_csv("boro_population.csv")
df_pop.head()
```

```python
combined_pop_df = df_pop.merge(combined_df, left_on="borough", right_on="borough")
combined_pop_df
```

```python
(combined_pop_df['borough'].value_counts() / combined_pop_df.groupby('borough')['population'].median()).sort_values(ascending=False)
```

```python
"Manhattan is the borough that has the highest number of dogs per-capita."
```
