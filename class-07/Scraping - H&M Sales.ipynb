## Request on-sale women's tops.

### How many results do you get?

```python
url_1 = 'http://api.hm.com/v2/us/en/products/categories?path=sale/ladies/tops&deviceType=DESKTOP'
response = requests.get(url_1)
```

```python
data_women_top = response.json()
data_women_top['displayArticles']
len(data_women_top['displayArticles'])
```

### How many total results are there? How do you change the URL to get more?

There are a few ways to change the URL to: there is one very easy way and one a-little-difficult way.

```python
# changed the pagesize to the number that can display all the available items on one page
url_2 = 'http://api.hm.com/v2/us/en/products/display?categories=sale%2Fladies%2Ftops&concealCategories=true&pageSize=60&page=1&deviceType=DESKTOP'
response = requests.get(url_2)
```

```python
data_women_top_2 = response.json()
len(data_women_top_2['displayArticles'])
```

### Download all on-sale women's tops and save them to a CSV.

I want this CSV to include

- Name
- Product code/number
- Old price
- On-sale price
- Item's URL

```python
data_women_top_2['displayArticles'][0].keys()
```

```python
women_top_info = data_women_top_2['displayArticles']
```

```python
for item in women_top_info:
    item['name']
```

```python
for item in women_top_info:
    item['productNumber']
```

```python
for item in women_top_info:
    item['priceInfo']['formattedOldPrice']
```

```python
for item in women_top_info:
    item['priceInfo']['price']
```

```python
for item in women_top_info:
    item['webUrl']
```

```python
women_top_list = []

for item in women_top_info:
    women_top_dic = {}
    
    print(item['name'])
    women_top_dic['name'] = item['name']
    
    print(item['productNumber'])
    women_top_dic['Product Number'] = item['productNumber']
    
    print(item['priceInfo']['formattedOldPrice'])
    women_top_dic['Old Price'] = item['priceInfo']['formattedOldPrice']
    
    print(item['priceInfo']['formattedPrice'])
    women_top_dic['On-sale Price'] = item['priceInfo']['formattedPrice']
    
    print(item['webUrl'])
    women_top_dic['URL'] = item['webUrl']
    
    print('---------')
    
    women_top_list.append(women_top_dic)
```

```python
women_top_list
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(women_top_list)
df.head()
```

```python
df.to_csv("women_top_list.csv", index=False)
```

## Analyzing on-sale items

### Read in the on-sale CSV and make sure the data looks right

```python
womentop_df = pd.read_csv("women_top_list.csv")
womentop_df.head()
```

### Convert the prices to numbers, if you haven't already

```python
womentop_df['Old Price'] = womentop_df['Old Price'].astype('float')
```

```python
womentop_df['On-sale Price'] = womentop_df['On-sale Price'].astype('float')
```

```python
womentop_df
```

### What is the average on-sale price of a women's top? Make a histogram, too, please.

```python
print('The average on-sale price of a women\'s top is $', round(womentop_df['On-sale Price'].mean()))
```

```python
%matplotlib inline
womentop_df['On-sale Price'].hist()
```

### What is the average discount between full price and on sale? Histogram too, please.

```python
discount = (womentop_df['Old Price'] - womentop_df['On-sale Price']).mean()
print('The average discount between full price and on sale is $', round(discount))
```

```python
discount = womentop_df['Old Price'] - womentop_df['On-sale Price']
discount.hist()
```

### What is the average discount as a percentage of the full price? Histogram too, please.

```python
average_discount = ((womentop_df['Old Price'] - womentop_df['On-sale Price']) *100 / womentop_df['Old Price']).mean()
print('The average discount as a percentage of the full price is', round(average_discount), '%.')
```

```python
discount = (womentop_df['Old Price'] - womentop_df['On-sale Price']) *100 / womentop_df['Old Price']
discount.hist()
```
