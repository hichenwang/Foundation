## Request the page with the correct data parameters AND the correct MINIMUM headers

This time it should work.

```python
headers = {
    'Referer':'https://www.tdlr.texas.gov/cimsfo/fosearch.asp',
    'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}
response = requests.post("https://www.tdlr.texas.gov/cimsfo/fosearch_results.asp", data=data, headers=headers)
```

```python
doc = BeautifulSoup(response.text, 'html.parser')
doc
```

## Scraping

### Loop through each `tr` and print each person's name

You'll get an error because the first one doesn't have a name. How do you make that not happen? I'm happy to help if you ask on the board.

```python
for barber in doc.find_all('tr'):
    barber_name = barber.find('span', attrs={"class":'results_text'})
    if barber_name:
        print(barber_name.text)
```

## Loop through each `tr`, printing each violation description

- TIP: What is the container tag name for it?
- TIP: You'll get an error even if you're ALMOST right - which row is causing the problem?

```python
for barber in doc.find_all('tr'):
    barber_violation = barber.find_all('td')
    if barber_violation:
        print(barber_violation[2].text)
```

## Loop through each `tr`, printing the complaint number

- TIP: It should be the last piece of the fist `td`

```python
for barber in doc.find_all('tr'):
    barber_complaint_number = barber.find('td')
    if barber_complaint_number:
        print(barber_complaint_number.find_all('span', attrs={"class":'results_text'})[-1].text)
```

## Saving the results

### Loop through each `tr` to create a list of dictionaries

Each dictionary must contain

- Person's name
- Violation description
- Violation number

Create a new dictionary for each `tr` (except the header).

```python
barber_violations = []

for barber in doc.find_all('tr')[1:]:
    current = {}
    barber_name = barber.find('span', attrs={"class":'results_text'})
    if barber_name:
        print(barber_name.text)
        current['barber_name'] = barber_name.text
    barber_violation = barber.find_all('td')
    if barber_violation:
        print(barber_violation[2].text)
        current['barber_violation'] = barber_violation[2].text
    barber_complaint_number = barber.find('td')
    if barber_complaint_number:
        print(barber_complaint_number.find_all('span', attrs={"class":'results_text'})[-1].text)
        current['barber_complaint_number'] = barber_complaint_number.find_all('span', attrs={"class":'results_text'})[-1].text
    barber_violations.append(current)
    print('-----')
```

```python
barber_violations
```

### Save that to a CSV

```python
import pandas as pd
```

```python
df = pd.DataFrame(barber_violations)
df.head()
```

```python
df.to_csv("barber_violations.csv", index=False)
```
